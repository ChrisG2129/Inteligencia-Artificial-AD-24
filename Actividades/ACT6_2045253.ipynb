```python
from re import X
import matplotlib.pyplot as plt #geerar graficos
import numpy as np#generar arreglos
import pandas as pd #caegar conjuntos de datos externos
import sklearn.linear_model #contiene diferentes modelos de
from sklearn.datasets import fetch_california_housing #conjunto de datos a tomar
from sklearn.model_selection import train_test_split #dividir conjuntos de datos
from sklearn.metrics import mean_squared_error #calcular el error cuadratico medio


#Cargar conjunto de datos
california = fetch_california_housing()
df = pd.DataFrame(data=california.data,columns=california.feature_names)
df['Precio'] = california.target

#Sekeccionar variable independiente
x = df[['MedInc']] #ingresos medios
y = df['Precio']

#Dividir datos de entrenamiento y prueba
train_data, test_data, train_labels, test_labels = train_test_split(x, y, test_size=0.2, random_state=42)

print(df)

#Para graficar
df.plot(kind='scatter', x="MedInc", y="Precio")
plt.show()

#Modelo
model = sklearn.linear_model.LinearRegression()
model.fit(train_data.values, train_labels) #Entrenamiento del modelo
predic = model.predict(test_data.values)

#Error cuadratico medio
mse = mean_squared_error(test_labels, predic)
print("Error cuadratico medio:", mse)

#Predicción del modelo
X_new = [[6.45025]] #Ingresos medios
print(model.predict(X_new))

X_new = [[10]] #Ingresos medios
print(model.predict(X_new))

X_new = [[3]] #Ingresos medios
print(model.predict(X_new))

#Graficar los resultados
plt.scatter(test_data, test_labels, color='blue', label='Datos')
plt.plot(test_data, predic, color='red', linewidth=2, label='Pedicción del Modelo')
plt.xlabel('Ingresos medios')
plt.ylabel('Precio')
plt.legend()
plt.show()
```
